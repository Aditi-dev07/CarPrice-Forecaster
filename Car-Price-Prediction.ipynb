# Task
Preprocess the car dataset by creating separate pipelines for numerical and categorical features, combining them using `ColumnTransformer`, and applying these transformations to the training and testing data.

## Create Preprocessing Pipelines

### Subtask:
Create separate pipelines for numerical and categorical features. The categorical pipeline will impute missing values with the most frequent value and then apply OneHotEncoder. The numerical pipeline will scale the data using StandardScaler.


**Reasoning**:
Based on the previous steps where categorical and numerical columns were identified, I will now define the `SimpleImputer` and `OneHotEncoder` for the categorical pipeline and combine them into `categorical_transformer`. Then, I will define `StandardScaler` for numerical features and create `numerical_transformer`. I will also filter out 'Price' from the numerical columns as it is the target variable.



````python
impute_categorical = SimpleImputer(strategy='most_frequent')
onehot_categorical = OneHotEncoder(handle_unknown='ignore')
categorical_transformer = Pipeline(steps=[('impute', impute_categorical), ('onehot', onehot_categorical)])

# Exclude 'Price' from numerical columns as it's the target variable
numerical_columns = [col for col in car_df.select_dtypes(include=['float', 'int']).columns if col != 'Price']
print("Numerical columns (excluding 'Price'): ")
print(numerical_columns)

scaler_numerical = StandardScaler()
numerical_transformer = Pipeline(steps=[('scale', scaler_numerical)])
print("Preprocessing pipelines for categorical and numerical features have been defined.")
````

## Combine Preprocessing Steps

### Subtask:
Combine the numerical and categorical preprocessing pipelines using a ColumnTransformer. This will apply the correct transformations to the respective columns in your dataset.


**Reasoning**:
To combine the numerical and categorical preprocessing pipelines, I will define a `ColumnTransformer` named `preprocessor`. This transformer will apply the `categorical_transformer` to the `categorical_columns` and the `numerical_transformer` to the `numerical_columns`, ensuring all columns are handled appropriately with `remainder='passthrough'`.



````python
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, ['Brand', 'Model', 'Fuel_Type', 'Transmission', 'Owner_Type']),
        ('num', numerical_transformer, ['Year', 'Kilometers_Driven', 'Mileage', 'Engine', 'Power', 'Seats'])
    ],
    remainder='passthrough'
)

print("ColumnTransformer 'preprocessor' has been successfully defined to combine the categorical and numerical pipelines.")
````

## Apply Preprocessing to Data

### Subtask:
Apply the created ColumnTransformer to the features (X_train and X_test) to preprocess the data.


**Reasoning**:
First, I need to separate the features and the target variable from the `car_df` and then split them into training and testing sets, as outlined in the instructions for preparing the data for preprocessing.



````python
features = car_df.drop(['Price'], axis=1)
label = car_df['Price']

X_train, X_test, y_train, y_test = train_test_split(features, label, random_state=42)
print("Features and target separated, and data split into training and testing sets.")
````

**Reasoning**:
Now that the data is split, I will apply the ColumnTransformer `preprocessor` to `X_train` using `fit_transform` to learn the transformations and apply them, and then apply it to `X_test` using `transform`.



````python
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)
print("Training and testing data have been preprocessed using the ColumnTransformer.")
````
